```python
import pandas as pd


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
```

```python
ds = pd.read_csv('house_price.csv')
```

```python
pd.reset_option('all')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 81)
```

```python
ds.isnull().sum()
```

```python
columns_to_encode = ds.select_dtypes(include=['object']).columns

# Initialize and apply LabelEncoder
label_encoder = LabelEncoder()
for col in columns_to_encode:
    ds[col] = label_encoder.fit_transform(ds[col])
```

```python
ds
```

```python
ds['Alley'] = ds['Alley'].fillna(ds['Alley'].mean())
ds['MasVnrType'] = ds['MasVnrType'].fillna(ds['MasVnrType'].mean())
ds['FireplaceQu'] = ds['FireplaceQu'].fillna(ds['FireplaceQu'].mean())
ds['PoolQC'] = ds['PoolQC'].fillna(ds['PoolQC'].mean())
ds['Fence'] = ds['Fence'].fillna(ds['Fence'].mean())
ds['MiscFeature'] = ds['MiscFeature'].fillna(ds['MiscFeature'].mean())
```

```python
ds.isnull().sum()
```

```python
ds = ds.dropna()
```

```python
ds.isnull().sum()
```

```python
x = ds.drop('SalePrice', axis = 1)
y = ds['SalePrice']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
# List of columns with missing values
columns_with_missing = [
    'MSZoning', 'LotFrontage', 'Alley', 'Utilities', 'Exterior1st', 'Exterior2nd', 
    'MasVnrType', 'MasVnrArea', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 
    'BsmtFinSF1', 'BsmtFinType2', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 
    'BsmtFullBath', 'BsmtHalfBath', 'KitchenQual', 'Functional', 'FireplaceQu', 
    'GarageType', 'GarageYrBlt', 'GarageFinish', 'GarageQual', 'GarageCond', 
    'SaleType', 'PoolQC', 'Fence', 'MiscFeature', 'GarageCars', 'GarageArea'
]

# Fill missing values with mean for numeric columns, mode for categorical columns
for col in columns_with_missing:
    if test[col].dtype in ['float64', 'int64']:  # Check if the column is numeric
        test[col] = test[col].fillna(test[col].mean())
    else:
        test[col] = test[col].fillna(test[col].mode()[0])  # Fill categorical columns with mode
```
